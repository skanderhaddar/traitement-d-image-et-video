```python
video = cv2.VideoCapture(r"C:\Users\USER\Desktop\tous les tp\tp traitement d'image\TP4_TRAIT_COD_Image_Video\Resources\Cars.mp4")
while (True):
	# stoquer l'image issue de la vidéo à l'instant t dans la variable "frame"
    ret, frame = video.read()
    # afficher l'image contenue dans "frame"
    cv2.imshow('output', frame)
	# quiter la boucle infinie lorqu'on appuie sur la touche 'q'
    if(cv2.waitKey(25) & 0xFF == ord('q')) :
        break
video.release()
cv2.destroyAllWindows()
```

```python
def extractImages(pathIn, pathOut):
    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    count = 0
    success = True
    while success:
      success,image = vidcap.read()
      print ('Read a new frame: ', success)
      cv2.imwrite( pathOut + "\\frame%d.jpg" % count, image)     # save frame as JPEG file
      count += 1
pathin = r"C:\Users\USER\Desktop\tous les tp\tp traitement d'image\TP4_TRAIT_COD_Image_Video\Resources\Cars.mp4"  
pathout = r"C:\Users\USER\Desktop\tous les tp\tp traitement d'image\TP4_TRAIT_COD_Image_Video\Resources\Cars2.mp4"
extractImages(pathin , pathout)
```
