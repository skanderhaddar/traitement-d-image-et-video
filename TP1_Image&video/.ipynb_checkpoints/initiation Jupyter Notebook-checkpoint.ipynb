### I.2. Types des données (images)

On trouve plusieurs conventions pour répresenter les valeurs des pixels d'une image:

```
  0 - 255   avec  0 est noir, 255 est blanc
  0 - 1     avec  0 est noir, 1 est blanc
```

Le type de l'image est déterminé par ``.dtype`` .


````python
# exemple 1
image1.dtype
````

On prend l'exemple du code ci-dessous permettant de générer deux images en niveaux de gris ayant deux différents types:

````python
# exemple 2

import numpy as np #importation de bibliothéque numpy 
from matplotlib import pyplot as plt #importation de bibliothéque matplotlib.pyplot 

image0 = np.linspace(0, 1, 900).reshape((30, 30))  #initialiser image0 900 valeurs entre 0 et 1 qui sont distribué d'une maniere equitable
                                                   #changer le shape de image0 vers (30,30)
image1 = np.linspace(0, 255, 900).reshape((30, 30)).astype(np.uint8)#initialiser image1 900 valeurs entre 0 et 255 qui sont distribué d'une maniere equitable
                                                                    #changer le shape de image1 vers (30,30)
                                                                    #changer le type de matrice vers int de 8 bits

print("image0:", image0.dtype, image0.min(), image0.max()) #afficher "image0:" , le type de image0, le minimum de image0 et le maximum de image0
print("image1:", image1.dtype, image1.min(), image1.max()) #afficher "image1:" , le type de image1, le minimum de image1 et le maximum de image1

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(5, 5)) #creation d'une figure de deux figures de taille (5,5)
ax0.imshow(image0 , cmap='gray')  #affichage d'image0
ax1.imshow(image1, cmap='gray') #affichage d'image1
plt.show()
````

 **Question 1:** Expliquer chaque commande dans le code précedent.

On peut changer le type des images en utilisant les deux représentations suivantes: `` img_as_float`` et `` img_as_ubyte``, etc. On prend l'exemple suivant: 

````python
from skimage import img_as_float, img_as_ubyte #importation de les fonction (img_as_float, img_as_ubyte) à partir skimag

from skimage import data #importation de bibilotheque data à partir skimag
 
image = data.chelsea() #initialiser une variage image qui contient l'image "chelsea" qui est dans la bibliotheque data
image1 = img_as_ubyte(image)  #convertir la matrice "image" en un autre matrice nommé "image1" qui est de type entier
image2 = img_as_float(image) #convertir la matrice "image" en un autre matrice nommé "image2" qui est de type reél

print("type, min, max:", image1.dtype, image1.min(), image1.max()) #affichage de "type , min , max:" , type de l'image1,son minimum et son maximum 
print("type, min, max:", image2.dtype, image2.min(), image2.max())#affichage de "type , min , max:" , type de l'image2,son minimum et son maximum 
print() #ajouter une ligne vide

print("max_image1/255 =", image1.max()/255.) #affichage de "max_image1/255 =" , le maximum de image1 diviser par 255
````

````python
#exercice 1

from skimage.color import rgb2gray
grayscale = rgb2gray(X)
print(type(grayscale))
print(grayscale.dtype)
print(grayscale.shape)
print(grayscale.min(), grayscale.max())
io.imshow(grayscale );
plt.imsave('images/lena_gray.jpg', grayscale , cmap="gray") # save image to disk
````

````python
#Exercice2
from skimage import io
lena_gray_inv= 255-grayscale
print(type(lena_gray_inv))
print(lena_gray_inv.dtype)
print(lena_gray_inv.shape)
print(lena_gray_inv.min(), lena_gray_inv.max())
plt.imshow(lena_gray_inv, cmap = "gray");
plt.imsave('images/lena_gray_inv.jpg', lena_gray_inv , cmap= 'gray') # save image to disk
````

````python
#Exerice 3
X = io.imread('images/lena.jpg')
plt.imshow(X);

fig, ax= plt.subplots(1,3)
ax[0].imshow(X[:,:,0], cmap= "Reds_r")
ax[0].set_title("image rouge ")
ax[1].imshow(X[:,:,1]  , cmap= "Greens_r")
ax[1].set_title("image vert ")
ax[2].imshow(X[:,:,2]  , cmap = "Blues_r")
ax[2].set_title("image bleu ")
plt.show()
plt.imsave('images/lena_red.jpg', X[:,:,0] , cmap= 'Reds') # save image to disk
plt.imsave('images/lena_green.jpg', X[:,:,1] , cmap= 'Greens_r') # save image to disk
plt.imsave('images/lena_bleu.jpg', X[:,:,2] , cmap= 'Blues_r') # save image to disk
````

````python
#Exercice 4

X = io.imread('images/lena.jpg')
from skimage.color import rgb2hsv
Lena_HSV = rgb2hsv(X)
hue_img = Lena_HSV[:, :, 0]
Saturation_img = Lena_HSV[:, :, 1]
value_img = Lena_HSV[:, :, 2]

fig, (ax0, ax1, ax2 , ax3) = plt.subplots(ncols=4, figsize=(8, 2))

ax0.imshow(X)
ax0.set_title("RGB image")
ax0.axis('off')
ax1.imshow(hue_img, cmap='gray')
ax1.set_title("Hue channel")
ax1.axis('off')
ax2.imshow(value_img ,cmap='gray')
ax2.set_title("Value channel")
ax2.axis('off')
ax3.imshow(Saturation_img,cmap='gray')
ax3.set_title("Saturation_img")
ax3.axis('off')


fig.tight_layout()
plt.imsave('images/Hue_Lena.jpg', X[:,:,0] , cmap= 'gray') # save image to disk
plt.imsave('images/Value_Lena.jpg', X[:,:,2] , cmap= 'gray') # save image to disk
plt.imsave('images/Saturation_Lena.jpg', X[:,:,1] , cmap= 'gray') # save image to disk
````

````python
#Exercice 6
from PIL import Image
def rajout_cadre(img1 , epsilon):
    im2=  np.random.random((X.shape[0]+epsilon*2 , X.shape[1]+epsilon*2))* 0
    plt.imsave('images/Black.jpg',im2 , cmap= 'gray')
    img2 = Image.open("images/Black.jpg")
    img1=Image.fromarray(img1)
    img2.paste(img1, ((epsilon,epsilon)))
    plt.imshow(img2);
    return(img2)
````

````python
img1 = io.imread('images/lena.jpg')

img2=rajout_cadre(img1 , 10)
````

````python
#exercice 5
def cadre_noir(img1,epsilon): 
    im2=  np.random.random((X.shape[0]+epsilon*2 , X.shape[1]+epsilon*2))* 0
    plt.imsave('images/Black.jpg',im2 , cmap= 'gray')
    img2 = Image.open("images/Black.jpg")
    shape= img1.shape
    img1=Image.fromarray(img1)
    img1 = img1.crop((epsilon, epsilon, shape[0]-epsilon,shape[1]-epsilon ))
    img1 = img1.resize((shape[0],shape[1]))

    img2.paste(img1, ((epsilon,epsilon)))
    plt.imshow(img2);
    return(img2)
````

````python
X = io.imread('images/lena.jpg')

new_img= cadre_noir(X,50)
````
